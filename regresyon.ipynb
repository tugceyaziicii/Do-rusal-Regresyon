```python
import pandas as pd
df=pd.read_csv("Advertising.csv")
df=df.iloc[:,1:len(df)]
df.head()
```

```python
df.info()
```

```python
import seaborn as sns
sns.jointplot(x="TV",y="sales",data=df,kind="reg")
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X=df[["TV"]]
X.head()
```

```python
y=df[["sales"]]
```

```python
reg= LinearRegression()
```

```python

model= reg.fit(X,y)
```

```python
model
```

```python
str(model)
```

```python
dir(model)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
#rkare bağımsız değişkenlerin bağımlı değişkenlerdeki verilerin yüzde kaçını açıkladığını anlatır
model.score(X,y)
```

```python
tahmin_edilen_y=pd.DataFrame(model.predict(X)[0:10])
```

```python
hatalar=pd.concat([gercek_y,tahmin_edilen_y],axis=1)
```

```python
hatalar.columns=["gercek_y","tahmin_edilen_y"]
```

```python
hatalar
```

```python
hatalar["hata"]=hatalar["gercek_y"]-hatalar["tahmin_edilen_y"]
```

```python
hatalar
```

```python
hatalar["hata_kareler"]=hatalar["hata"]**2
```
